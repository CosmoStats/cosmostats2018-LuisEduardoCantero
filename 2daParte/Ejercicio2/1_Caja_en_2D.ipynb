## a) Puntos en una malla cuadriculada de distancia (MxM cada retícula)

Calculamos algunos parametros que nos serviran para construir la caja. Los puntos no deben estar en el borde de la caja.

```python
separacion = 25
longitud = 1000
```

Generamos la distribucion de puntos.

```python
def construye (l, s):
    nxlado = int(l / s)
    ntot = nxlado ** 2
    px = []
    py = []
    c = 1
    for i in range(1, nxlado):
        for j in range(1, nxlado):
            px.append(separacion * i)
            py.append(separacion * j)
            c += 1
    return px, py        
```

```python
x, y = construye(longitud, separacion)
```

```python
q1 = np.linspace(0, 0, 2)
q2 = np.linspace(0, longitud, 2)
q3 = np.linspace(longitud, longitud, 2)

plt.figure(figsize=(7, 7))
plt.plot(q1, q2, color = 'g')
plt.plot(q3, q2, color = 'g')
plt.plot(q2, q3, color = 'g')
plt.plot(q2, q1, color = 'g')

plt.plot(x,y, "bo")
```

## b) 1000 puntos sobre un disco de radio 300.

```python
radio = 300
n = 1000

xx1 = []
yy1 = []

j = 0
u = 0
while u < (n):
    fx = np.random.uniform(-radio, radio)
    fy = np.random.uniform(-radio, radio)
    rad = math.sqrt(math.pow(fx, 2) + math.pow(fy, 2))
    if rad <= radio:
        xx1.append(fx)
        yy1.append(fy) #Faltan puntos
        u += 1
    j += 1
```

```python
len(xx1)
```

```python
plt.figure(figsize=(7, 7))
plt.plot(xx1,yy1, "*")
```

## c) 1000 puntos sobre un anillo con radio mayor 300 y radio menor 290

```python
radio = 300
radioint = 290
n = 1000

xx2 = []
yy2 = []

j = 0
u = 0
while u < (n):
    fx = np.random.uniform(-radio, radio)
    fy = np.random.uniform(-radio, radio)
    rad = math.sqrt(math.pow(fx, 2) + math.pow(fy, 2))
    if rad <= radio:
        if rad >= radioint:
            xx2.append(fx)
            yy2.append(fy) #Faltan puntos
            u += 1
    j += 1
```

```python
len(xx2)
```

```python
plt.figure(figsize=(7, 7))
plt.plot(xx2,yy2, "*")
```
