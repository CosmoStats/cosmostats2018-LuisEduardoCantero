# Como muestrear una PDF

## Generamos valores para la funcion acumulada de la Distribucion de Gauss

```python
mu1 = 70
sig1 = 2
h = 0.01
```

```python
x1 = np.arange(60, 80, h)
a1 = np.empty(len(x1))
a2 = np.empty(len(x1))
y1 = np.empty(len(x1))
```

```python
for i in range(0, len(x1)):
    a1[i] = 0.5 * (1 + math.erf((x1[i] - mu1)/(sig1*math.sqrt(2))))
```

```python
plt.plot(x1,a1, 'black', label="CDF NORMAL $μ = 70, σ = 2$")
plt.legend()
plt.xlabel("$H_{0}$")
plt.ylabel("$F(H_{0})$")
```

```python
for j in range(0, len(x1)):
    a2[j] = mu1 + sig1*math.sqrt(2)* erfinv(2*a1[j]-1)

    
```

```python
plt.plot(x1,a2, 'black', label="PDF NORMAL NORMAL $μ = 70, σ = 2$")
plt.legend()
plt.xlabel("$H_{0}$")
plt.ylabel("$P(H_{0})$")
```

```python
for j in range(1, len(x1 - 1)):
    y1[j] = (a1[j] - a1[j-1])/(h)    
```

```python
plt.plot(x1,y1, 'black', label="PDF NORMAL NORMAL $μ = 70, σ = 2$")
plt.legend()
plt.xlabel("$H_{0}$")
plt.ylabel("$P(H_{0})$")
```
