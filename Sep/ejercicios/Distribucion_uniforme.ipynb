# DISTRIBUCION DE PUNTOS UNIFORME EN UNA CIRCUNFERENCIA

```python
# Aqui cargamos las liberias a usar

import matplotlib.pyplot as plt
import scipy.special
import numpy as np
import math
%matplotlib inline
```

```python
tam = 800
fx1 = np.random.rand(tam, 2)
x1 = np.empty(tam)
y1 = np.empty(tam)
n1 = np.empty(tam)
m1 = np.empty(tam)
r = np.empty(tam)
```

```python
for j in range(0, tam):
    r[j] = math.sqrt(math.pow(fx1[j,0],2)+math.pow(fx1[j,1],2))
    if r[j] <= 1:
                 x1[j] = fx1[j, 0]
                 y1[j] = fx1[j, 1]
                 n1[j] = 0
                 m1[j] = 0
    else:
                 n1[j] = fx1[j, 0]
                 m1[j] = fx1[j, 1]
                 x1[j] = 0
                 y1[j] = 0
                 
```

```python
plt.plot(x1,y1, "*")
plt.plot(n1,m1, "*")
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
```
